```python
import gradio as gr
import heapq
from collections import deque
import random
```

```python
rules = {

    # Healthcare
    "Fever": ("Healthcare", "Possible Viral Infection"),
    "Cough": ("Healthcare", "Possible Respiratory Infection"),
    "Headache": ("Healthcare", "Possible Stress or Migraine"),

    # Education
    "Low Attendance": ("Education", "Student May Fail Internal Assessment"),
    "High Marks": ("Education", "Excellent Academic Performance"),

    # Banking
    "Multiple Failed Transactions": ("Banking", "Possible Fraud Detection"),
    "High Credit Score": ("Banking", "Eligible for Loan Approval"),

    # Cyber Security
    "Multiple Login Attempts": ("Cyber Security", "Possible Brute Force Attack"),
    "Unknown Device Login": ("Cyber Security", "Suspicious Login Activity"),

    # E-Commerce
    "Repeated Product Search": ("E-Commerce", "Recommend Similar Products"),
    "High Purchase History": ("E-Commerce", "Premium Customer Recommendation")
}
```

```python
graph = {
    "Input": ["Knowledge Base"],
    "Knowledge Base": ["Reasoning Engine"],
    "Reasoning Engine": ["Explanation Generator"],
    "Explanation Generator": ["Output"],
    "Output": []
}

def bfs(start):

    visited = []
    q = deque([start])

    while q:

        node = q.popleft()

        if node not in visited:
            visited.append(node)

            for i in graph[node]:
                q.append(i)

    return visited
```

```python
astar_graph = {
    "A": [("B", 1), ("C", 3)],
    "B": [("D", 2)],
    "C": [("D", 1)],
    "D": []
}

heuristic = {
    "A": 3,
    "B": 2,
    "C": 1,
    "D": 0
}

def astar(start, goal):

    pq = [(0, start)]
    visited = []

    while pq:

        cost, node = heapq.heappop(pq)

        if node == goal:
            visited.append(node)
            break

        if node not in visited:

            visited.append(node)

            for neigh, w in astar_graph[node]:
                heapq.heappush(
                    pq,
                    (cost + w + heuristic[neigh],
                     neigh)
                )

    return visited
```

```python
colors = ["Red", "Green", "Blue"]

states = {
    "A": None,
    "B": None,
    "C": None
}

adj = {
    "A": ["B", "C"],
    "B": ["A", "C"],
    "C": ["A", "B"]
}

def solve_csp():

    temp = states.copy()

    for s in temp:

        for c in colors:

            valid = True

            for n in adj[s]:
                if temp[n] == c:
                    valid = False

            if valid:
                temp[s] = c
                break

    return temp
```

```python
def utility_function(score):

    if score > 80:
        return "High Utility"

    elif score > 50:
        return "Medium Utility"

    else:
        return "Low Utility"
```

```python
def bayes():

    p_a = 0.10
    p_b_given_a = 0.80
    p_b = 0.20

    return round(
        (p_b_given_a * p_a) / p_b,
        2
    )
```

```python
def markov():

    state = "Sunny"

    transition = {
        "Sunny": "Rainy",
        "Rainy": "Sunny"
    }

    for i in range(5):
        state = transition[state]

    return state
```

```python
def explainable_ai(user_input):

    # Backend execution
    bfs("Input")
    astar("A", "D")
    solve_csp()
    utility_function(85)
    bayes()
    markov()

    # Rule-based inference
    if user_input in rules:

        domain, prediction = rules[user_input]

    else:

        domain = "Unknown"
        prediction = "No Matching Rule Found"

    html = f"""

    <div style="
    max-width:950px;
    margin:auto;
    padding:40px;
    border-radius:25px;
    background:linear-gradient(135deg,#0f172a,#1e3a8a);
    font-family:Arial;
    color:white;
    ">

        <h1 style="
        text-align:center;
        color:#facc15;">
        Explainable AI Reasoning Engine
        </h1>

        <br>

        <div style="
        background:white;
        color:black;
        padding:25px;
        border-radius:20px;
        margin-bottom:20px;
        ">

            <h2 style="color:#2563eb;">
            Prediction Result
            </h2>

            <p><b>Selected Input :</b> {user_input}</p>

            <p><b>Domain :</b> {domain}</p>

            <p><b>AI Prediction :</b> {prediction}</p>

        </div>

        <div style="
        background:white;
        color:black;
        padding:25px;
        border-radius:20px;
        ">

            <h2 style="color:#16a34a;">
            Explanation
            </h2>

            <p style="
            line-height:1.8;
            font-size:18px;
            ">

            The system analyzed the selected
            input and compared it with the
            knowledge base. Intelligent reasoning
            techniques were applied to generate
            the prediction shown above.

            The explanation helps users understand
            why the decision was made and makes
            the Artificial Intelligence system
            transparent, reliable, and easy to
            understand.

            </p>

        </div>

        <br>

        <center>
        <h3 style="color:#facc15;">
        Multi-Domain Explainable Artificial
        Intelligence System
        </h3>
        </center>

    </div>

    """

    return html
```

```python
demo = gr.Interface(

    fn=explainable_ai,

    inputs=gr.Dropdown(

        [
            "Fever",
            "Cough",
            "Headache",
            "Low Attendance",
            "High Marks",
            "Multiple Failed Transactions",
            "High Credit Score",
            "Multiple Login Attempts",
            "Unknown Device Login",
            "Repeated Product Search",
            "High Purchase History"
        ],

        label="Select Input"
    ),

    outputs=gr.HTML(),

    title="Explainable AI Reasoning Engine",

    description="""
Select an input and the system will
generate an AI prediction along with
a simple explanation.
"""
)

demo.launch(share=True, debug=True)
```
